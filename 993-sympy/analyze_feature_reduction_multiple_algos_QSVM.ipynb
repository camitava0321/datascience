```python
# Feature Extraction with RFE
import pandas as pd
from pandas import read_csv
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
# load data
# url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv"
# names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df = read_csv("diab_data.csv")
dataframe = df.iloc[:50] 

dataframe.count()
```

```python
array = dataframe.values
X = array[:,0:8]
Y = array[:,8]
print("X :", X)
print( "Y: ", Y)
```

```python
num_features = 8
# QUANTUM FEATURE MAP
reps = 3
feature_map = ZZFeatureMap(feature_dimension=num_features, reps=reps, entanglement='linear')
# QUANTUM INSTANCE BACKEND
backend = BasicAer.get_backend('statevector_simulator')
# statevector_sim is single shot so much faster for smaller problems
quantum_instance = QuantumInstance(backend)
qsvm = QSVM(feature_map)

def quantum_kernel(X, Y):
    return qsvm.get_kernel_matrix(quantum_instance, feature_map, X, Y)


feature_map.draw()
```

# CLASSICAL RFE 

```python
estimator = SVR(kernel="linear")
selector = RFE(estimator, n_features_to_select=2, step=1)
selector.fit(X, Y)
print("Num Features: %d" % selector.n_features_)
print("Selected Features: %s" % selector.support_)
print("Feature Ranking: %s" % selector.ranking_)
```

```python
selector.estimator_
```

# CLASSICAL RFECV


```python
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.svm import SVC

svc = SVC(kernel="linear")

min_features_to_select = 2  # Minimum number of features to consider
rfecv = RFECV(estimator=svc, step=1, cv=StratifiedKFold(2),
              scoring='accuracy',
              min_features_to_select=min_features_to_select)
rfecv.fit(X, Y)
print("Optimal number of features : %d" % rfecv.n_features_)
```

# QUANTUM RFECV 

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV


# The "accuracy" scoring is proportional to the number of correct
# classifications
rfecv_estimator_with_quantum_kernel = SVC(kernel=quantum_kernel)
min_features_to_select = 3  # Minimum number of features to consider
q_rfecv = RFECV(estimator=rfecv_estimator_with_quantum_kernel, step=1, cv=StratifiedKFold(2),
              scoring='accuracy',
              min_features_to_select=min_features_to_select)
print( "rfecv_estimator_with_quantum_kernel " , rfecv_estimator_with_quantum_kernel) 
# q_rfecv.fit(X, Y)
```

# CLASSICAL  SFS

```python
from sklearn.feature_selection import SequentialFeatureSelector
SVR_estimator = SVR(kernel="linear")


sfs = SequentialFeatureSelector(SVR_estimator, n_features_to_select=3, direction='backward')
sfstrained = sfs.fit(X, Y)
print( "selected features : ", sfstrained.get_support())
print( "sfstrained.get_params() : ", sfstrained.get_params())
```

# Quantum SFS

```python
Q_SVR_estimator = SVR(kernel=quantum_kernel)

print("Q_SVR_estimator : ", Q_SVR_estimator )

q_sfs = SequentialFeatureSelector(Q_SVR_estimator,  direction='backward', n_features_to_select=3)
print("q_sfs : ", q_sfs )

sfs1 = q_sfs.fit(X, Y)
print( "selected features : ", sfs1.get_support())
print( "sfs1.get_params() : ", sfs1.get_params())
```
